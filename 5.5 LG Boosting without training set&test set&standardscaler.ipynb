```python
#importing the Libraries#
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset#
dataset=pd.read_csv('insurance_pre.csv')
```

```python
#what saved in this file#
dataset
```

```python
#create dummy data set. #drop a column#
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#what saved in this file#
dataset
```

```python
#catogorical data convert to numerical data#
dataset=dataset.astype(int)
```

```python
#what saved in this file#
dataset
```

```python
#show all columns in dataset#
dataset.columns
```

```python
#independent variable means input.  #split input variables. #dataset assigned to independent variable#
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
#what saved in this file#
independent
```

```python
#dependent variable means output.  #split output variables. #dataset assigned to dependent variable#
dependent=dataset[['charges']]
```

```python
#what saved in this file#
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor
```

```python
#Model creation#
from sklearn.model_selection import GridSearchCV

# Set the objective directly in the model
model = LGBMRegressor()  

param_grid = {
    'boosting_type': ['gbdt'],
    'n_estimators': [10, 100, 1000,2000,3000],
     'learning_rate': [0.01, 0.05, 0.1],
    'num_leaves': [31, 50, 100],
    'max_depth': [-1, 5, 10],
    'min_data_in_leaf': [20, 50, 100],
    'importance_type': ['split', 'gain']
}


grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1)
grid_search.fit(independent,dependent)
```

```python
#Print best parameter afterr tuning#
print("Best parameters found:", grid_search.best_params_)
print("Best score (neg_mean_squared_error):", grid_search.best_score_)
```

```python
# Extract the cross-validation results
cv_results = grid_search.cv_results_
```

```python
# Create a DataFrame from the cross-validation results
results_df = pd.DataFrame(cv_results)
```

```python
# Display the top rows of the results DataFrame
print(results_df.head())
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("children:"))
sex_male_input=int(input("sex Male 0 or 1:"))
smoker_yes_input=int(input("smoker yes 0 or 1:"))
```

```python
Future_Prediction=grid_search.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
