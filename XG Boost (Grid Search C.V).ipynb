```python
#importing the Libraries#
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset#
dataset=pd.read_csv('insurance_pre.csv')
```

```python
#what saved in this file#
dataset
```

```python
#create dummy data set. #drop a column#
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#what saved in this file#
dataset
```

```python
#catogorical data convert to numerical data#
dataset=dataset.astype(int)
```

```python
#what saved in this file#
dataset
```

```python
#show all columns in dataset#
dataset.columns
```

```python
#independent variable means input.  #split input variables. #dataset assigned to independent variable#
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
#what saved in this file#
independent
```

```python
#dependent variable means output.  #split output variables. #dataset assigned to dependent variable#
dependent=dataset[['charges']]
```

```python
#what saved in this file#
dependent
```

```python
#machine learning algorithms provider is sklearn #split train set and test set#
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
#Model creation#
from sklearn.model_selection import GridSearchCV
param_grid = {'loss':['squared_error','absolute_error','huber','quantile'],
              'max_features':[10,100,1000,2000,3000],'criterion':['friedman_mse','squared_error']}

grid=GridSearchCV(GradientBoostingRegressor(), param_grid, refit=True, verbose=3,n_jobs=-1)

#fitting the model for grid search
grid.fit(independent,dependent)
```

```python
#Print best parameter afterr tuning#
#Print(grid.best_params_)

re=grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
