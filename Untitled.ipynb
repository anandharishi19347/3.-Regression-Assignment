```python
#importing the Libraries#
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset#
dataset=pd.read_csv('insurance_pre.csv')
```

```python
#what saved in this file#
dataset
```

```python
#create dummy data set. #drop a column#
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#what saved in this file#
dataset
```

```python
#catogorical data convert to numerical data#
dataset=dataset.astype(int)
```

```python
#what saved in this file#
dataset
```

```python
#show all columns in dataset#
dataset.columns
```

```python
#independent variable means input.  #split input variables. #dataset assigned to independent variable#
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
#what saved in this file#
independent
```

```python
#dependent variable means output.  #split output variables. #dataset assigned to dependent variable#
dependent=dataset[['charges']]
```

```python
#what saved in this file#
dependent
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
#from sklearn.tree import Random Forest

param_grid = {'criterion':['poisson', 'absolute_error', 'squared_error', 'friedman_mse'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]}


grid_search= GridSearchCV(RandomForestRegressor(), param_grid=param_grid, refit = True, verbose= 3,n_jobs=-1)
# fitting the model for grid search
grid_search.fit(independent,dependent)
```

```python
# print best parameter after tuning
#print(grid.best_params_)
re=grid_search.cv_results_
print("The R_score value for best parameter {}:".format(grid_search.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
